```python
import urllib.request
from urllib.parse import urlencode
import os
import pandas as pd

token = os.environ['EPC_TOKEN']
output_folder = "epc_test/"
query_size = 5000

def download_data(local_authority_code, local_authority_name):

    base_url = 'https://epc.opendatacommunities.org/api/v1/domestic/search'
    query_params = {
        'size': query_size,
        'local-authority': local_authority_code,
        'from-year': 2010
        # 'from-month': 5, 'to-year':2015, ;'to-month': 10 
    }
    
    headers = {
    	'Accept': 'text/csv',
    	'Authorization': f'Basic {token}'
    }
    
    part_n = 0
    first_request = True
    next_page = None
    while next_page != None or first_request:
        if not first_request:
            query_params["search-after"] = next_page
        encoded_params = urlencode(query_params)
        url = f"{base_url}?{encoded_params}"
        response = urllib.request.urlopen(urllib.request.Request(url, headers=headers))
        response_body = response.read()
        data = response_body.decode()    
        next_page = response.getheader('X-Next-Search-After')
        part_n+= 1
        output_file = f'{output_folder}{local_authority_name}_{part_n}.csv'
        with open(output_file, "w") as f:
            f.write(data)
        first_request = False

local_authorities = pd.read_csv("../../../../data/london-local-authorities.csv")

for i in local_authorities.index:
    local_authority_code = local_authorities['Code'][i]
    local_authority_name = local_authorities['Description'][i]
    print(f"started {i}")
    download_data(local_authority_code, local_authority_name)
    print(f"completed {i}")
```
